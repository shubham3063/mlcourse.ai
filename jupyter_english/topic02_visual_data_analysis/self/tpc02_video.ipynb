## tSNE: t-distributed Stochastic Neighbour Embedding
- If there two points close in original d-dimensional space, then in 2D it should be close together and vice versa
- This is a trial for intuitive understanding of data distribution in d-dim space
- Algorithm is less than O(l^2)

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# to get documentation of any function / module
# TSNE?
```

```python
X = df.drop(['Churn','State'],axis=1,inplace=False) # this returns the new dataframe, not a reference!
```

```python
X.head()
```

```python
scaler = StandardScaler()
Xscaled = scaler.fit_transform(X)
```

```python
Xscaled.shape
```

```python
#%%time
tsne = TSNE(random_state=17)
tsneFit = tsne.fit_transform(Xscaled)
```

```python
tsneFit.shape
```

```python
plt.scatter(tsneFit[:,0], tsneFit[:,1], alpha=0.5);
```

```python
srsChrn = df['Churn'] == 1
srsLyl  = df['Churn'] == 0
```

```python
plt.scatter(tsneFit[srsChrn,0], tsneFit[srsChrn,1], alpha=0.5, c='blue', label='Churn')
plt.scatter(tsneFit[srsLyl,0], tsneFit[srsLyl,1], alpha=0.5, c='orange', label='Loyal')
plt.xlabel('tsne axis 1')
plt.ylabel('tsne axis 2')
plt.title('TSNE Representation');
plt.savefig('churn_tsne.png', dpi=300)
```
